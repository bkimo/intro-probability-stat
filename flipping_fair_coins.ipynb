### Fliping a Fair Coin Thousands of Times

```python
flips = coinflip.flip_fair_coins(100)
```

Let's plot how many times we see the two possible outcomes in a histogram

```python
coinflip.plot_discrete_histogram(flips)
```

Often what we will care about in this course is the fraction (also called the **frequency**) of times an outcome happens. To plot the fraction of times heads or tails occurred, we again use the `plot_discrete_histogram` function but now add the keyword argument `frequency=True`:

```python
coinflip.plot_discrete_histogram(flips, frequency=True)
```

Next, let's plot the fraction of heads as a function of the number of flips (going up to 100,000 flips).

```python
n = 100000
heads_so_far = 0
fraction_of_heads = []
for i in range(n):
    if coinflip.flip_fair_coin() == 'heads':
        heads_so_far += 1
    fraction_of_heads.append(heads_so_far / (i+1))
```

Note that `fraction_of_heads[i]` tells us what the fraction of heads is after the first i tosses. Then to actually plot the fraction of heads vs the number of tosses, enter the following:

```python
plt.figure(figsize=(8,6))
plt.axis([0, n, 0, 1])
plt.yticks(np.linspace(0, 1, 9, endpoint=True))
plt.plot(range(1,n+1), fraction_of_heads)
plt.xlabel('Figure 1. Number of flips')
plt.ylabel('Fraction of heads')
```

The fraction of heads initially can be far from 1/2 but as the number of flips increases, the fraction stabilizes and gets closer to 1/2, the probability of heads.

```python
n = 100000
heads_so_far = 0
fraction_of_heads = []
for i in range(n):
    if coinflip.flip_fair_coin() == 'heads':
        heads_so_far += 1
    fraction_of_heads.append(heads_so_far / (i+1))

plt.figure(figsize=(8,6))
plt.axis([0, n, 0, 1])
plt.yticks(np.linspace(0, 1, 9, endpoint=True))
plt.plot(range(1,n+1), fraction_of_heads)
plt.xlabel('Figure 2. Number of flips')
plt.ylabel('Fraction of heads')
```

The graphs in Figures 1 and Figure 2 show that in the short term (fewer flips) the observed proportion of heads is different and unpredictable for each experiment. As the number of flips increases, however, both graphs tend toward a proportion of $0.5$. This is the basic premise of probability. **Probability** is the *measure of the likelihood of a random phenomenon or chance behavior occurring*. It deals with experiments that yield random short-term results or **outcomes** yet reveal long-term predictability. **The long-term proportion in which a certain outcome is observed is the probability of that outcome**.

## The Law of Large Numbers

So we say that the probability of observing a head is $\frac{1}{2}$or $50\%$ or $0.5$ because, as we flip
the coin more times, the proportion of heads tends toward $\frac{1}{2}$. This phenomenon, which is illustrated in Figure 1 and 2, is referred to as the [Law of Large Numbers](https://en.wikipedia.org/wiki/Law_of_large_numbers).
